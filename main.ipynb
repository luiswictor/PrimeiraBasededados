# Quantidade de ingressantes no curso 

```python
qtd_ingressantes_curso = base_norte_nordeste[["nome_do_curso_de_graduacao",'qtd_ingressantes_curso']].groupby('nome_do_curso_de_graduacao').sum()
qtd_ingressantes_curso
```

```python
total_ingressantes = base_norte_nordeste['qtd_ingressantes_curso'].sum()
total_ingressantes
```

```python
qtd_ingressantes_curso['%'] = qtd_ingressantes_curso['qtd_ingressantes_curso'] / total_ingressantes * 100
qtd_ingressantes_curso
```

# Quantidade de permanentes no curso

```python
qtd_permanentes_curso = base_norte_nordeste[["nome_do_curso_de_graduacao",'qtd_perrmanencia_curso_ano_referencia']].groupby('nome_do_curso_de_graduacao').sum()
qtd_permanentes_curso
```

# Quantidade de concluintes 

```python
qtd_concluintes_curso = base_norte_nordeste[["nome_do_curso_de_graduacao",'qtd_concluintes_curso']].groupby('nome_do_curso_de_graduacao').sum()
qtd_concluintes_curso
```

# Quantidade de Desistentes

```python
qtd_desistentes_curso = base_norte_nordeste[["nome_do_curso_de_graduacao",'qtd_desistencia_no_curso']].groupby('nome_do_curso_de_graduacao').sum()
qtd_desistentes_curso
```

# Quantidade de Falecidos

```python
qtd_falecidos_curso = base_norte_nordeste[["nome_do_curso_de_graduacao",'qtd_falecimentos']].groupby('nome_do_curso_de_graduacao').sum()
qtd_falecidos_curso
```

# Base relacionando apenas cursos

```python
base_1 = qtd_ingressantes_curso.merge(qtd_permanentes_curso,on='nome_do_curso_de_graduacao',how='outer')
base_1
```

```python
base_2 = base_1.merge(qtd_concluintes_curso,on='nome_do_curso_de_graduacao',how='outer')
base_3 = base_2.merge(qtd_desistentes_curso,on='nome_do_curso_de_graduacao',how='outer')
base_final = base_3.merge(qtd_falecidos_curso,on='nome_do_curso_de_graduacao',how='outer')
base_final
```
